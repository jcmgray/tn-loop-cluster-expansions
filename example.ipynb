These example shows how to generate a simple update PEPS, compute its energy
using boundary contraction, then compute the energy using the loop cluster
expansion. All results are cached to disk in the `data` directory, delete 
entries or set `overwrite=True` to recompute.

```python
import simpclustlib
```

```python
model = "fermi_hubbard_U8"
geom = "square_obc"
```

Generate the simple update (SU) wavefunction (note possible symmetries and
site charges will depend on the model and geometry):

```python
kwargs = {
    "model": model,
    "geom": geom,
    "L": 4,
    "D": 4,
    "schedule": "D",  # how to set imaginary time step
    "init": "D-1",
    "seed": 0,
    "symm": "U1",
    "site_charge": "uniform-1",  # implies half-filling
}
```

```python
simpclustlib.memoize_groundstate_simple_update(
    progbar=True,
    **kwargs,
)
```

Compute reference boundary energy (for square_obc only):

```python
simpclustlib.memoize_energy_full(
    mode="mps",
    chi=32,
    progbar=True,
    **kwargs,
)
```

```python
simpclustlib.memoize_energy_gloop_expand(
    max_size=6,
    grow_from="alldangle",
    optimize="random-greedy-128",
    progbar=True,
    **kwargs,
)
```
